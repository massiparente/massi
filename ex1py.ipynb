```python
fname="data1.txt"
u1=[]
u2=[]
with open(fname,"r") as f:
    for line in f:
        rows = line.split(" ")
        u1.append ( float (rows[1]))
        u2.append (float (rows[2]))

fname="data2.txt"
u3=[]
u4=[]
with open(fname,"r") as f:
    for line in f:
        rows = line.split(" ")
        u3.append ( float (rows[1]))
        u4.append (float (rows[2]))
```

```python
u=np.zeros([4,len(u1)])
u[0]=u1
u[1]=u2
u[2]=u3
u[3]=u4
```

```python
block=np.zeros([4,15,len(u1)])
block_media=np.zeros([4,15])
block_var=np.zeros([4,15])
block_sigma=np.zeros([4,15])
```

```python
block.shape
```

```python
N=len(u1)

for i in range(4):
    media=0
    var=0
    dim=int(len(u1)/2)
    for k in range(int(len(u1)/2)):
    
        block[i][1][k]=0.5*(u[i][2*k+1]+u[i][2*k])
        media+=block[i][1][k]
        var+=block[i][1][k]**2
        block_media[i][1]=media/(int(len(u1)/2))
        block_var[i][1]=2*(var/dim-block_media[i][1]**2)
        block_sigma[i][1]=(block_var[i][1]/N)**0.5

```

```python
for i in range(4):
    for j in range(2,14):
        media=0
        var=0
        dim=int(len(u1)/(2**(j)))
        print(dim)
        for k in range(dim):
            #print(j)
            block[i][j][k]=0.5*(block[i][j-1][(2*k)+1]+block[i][j-1][2*k])
            media+=block[i][j][k]
            var+=block[i][j][k]**2
            block_media[i][j]=media/dim
            block_var[i][j]=(2**j)*(var/dim-block_media[i][j]**2)
            block_sigma[i][j]=(block_var[i][j]/N)**0.5
```
